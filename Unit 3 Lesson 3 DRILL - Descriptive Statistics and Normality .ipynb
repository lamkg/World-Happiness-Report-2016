```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
rand_exp = np.random.exponential(15, 100)

plt.hist(rand_exp)

plt.axvline(rand_exp.mean(), color='c', linestyle='solid', linewidth=2)

plt.axvline(rand_exp.mean() + rand_exp.std(), color='c', linestyle='dashed', linewidth=2)

plt.axvline(rand_exp.mean() - rand_exp.std(), color='c', linestyle='dashed', linewidth=2)

plt.show()
```

```python
rand_geo = np.random.geometric(.2, 100)

plt.hist(rand_geo)

plt.axvline(rand_geo.mean(), color='c', linestyle='solid', linewidth=2)

plt.axvline(rand_geo.mean() + rand_geo.std(), color='c', linestyle='dashed', linewidth=2)

plt.axvline(rand_geo.mean() - rand_geo.std(), color='c', linestyle='dashed', linewidth=2)


plt.show()
```

```python
rand_power = np.random.power(2, 100)

plt.hist(rand_power, bins=12)

plt.axvline(rand_power.mean(), color='c', linestyle='solid', linewidth=2)

plt.axvline(rand_power.mean() + rand_power.std(), color='c', linestyle='dashed', linewidth=2)

plt.axvline(rand_power.mean() - rand_power.std(), color='c', linestyle='dashed', linewidth=2)



plt.show()
```

```python
rand_par = np.random.pareto(12, 100)

plt.hist(rand_par, bins=12)

plt.axvline(rand_par.mean(), color='c', linestyle='solid', linewidth=2)

plt.axvline(rand_par.mean() + rand_par.std(), color='c', linestyle='dashed', linewidth=2)

plt.axvline(rand_par.mean() - rand_par.std(), color='c', linestyle='dashed', linewidth=2)


plt.show()
```

```python
rand_stand_exp = np.random.standard_exponential(100)

plt.hist(rand_stand_exp)

plt.axvline(rand_stand_exp.mean(), color='c', linestyle='solid', linewidth=2)

plt.axvline(rand_stand_exp.mean() + rand_stand_exp.std(), color='c', linestyle='dashed', linewidth=2)

plt.axvline(rand_stand_exp.mean() - rand_stand_exp.std(), color='c', linestyle='dashed', linewidth=2)

plt.show()
```

```python
rand_stand_t = np.random.standard_t(10, 100)

plt.hist(rand_stand_t)

plt.axvline(rand_stand_t.mean(), color='c', linestyle='solid', linewidth=2)

plt.axvline(rand_stand_t.mean() + rand_stand_t.std(), color='c', linestyle='dashed', linewidth=2)

plt.axvline(rand_stand_t.mean() - rand_stand_t.std(), color='c', linestyle='dashed', linewidth=2)

plt.show()
```

For data, that has a large x range, it would throw off the mean a bit, the line would be where the main clusters are, but if you removed the outlier the line would be more accurate. The standard deviation is a bit more useful when it's used against a normal distribution. I guess it didn't really help that I chose exponential curves that wouldn't have normal distributions. All data no matter what graph shows a bit of skew, whether be it a large or small skew.

```python
var1 = np.random.normal(5,.5, 100)

plt.hist(var1)

plt.show()
```

```python
var2 = np.random.normal(10, 1, 100)

plt.hist(var2)

plt.show()
```

```python
var1 = np.random.normal(5,.5, 100)
var2 = np.random.normal(10,1, 100)
var3 = var1+var2

mean = np.mean(var3)
sd = np.std(var3)

plt.hist(var3)
plt.axvline(x=mean,color='c')
plt.axvline(x=mean+sd, color='c', linestyle='dashed')
plt.axvline(x=mean-sd, color='c', linestyle='dashed')
plt.show()
```

5. Results will vary depending on the distribution of the random function and there should be a bit of a skew in all graphs.
